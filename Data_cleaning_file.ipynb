```python
# Converting the columns ['stars', 'followers', 'repositories'] to numeric data types
# Non-numeric values are coerced into NaN and replaced with 0 to handle inconsistencies
columns_to_fix = ['stars', 'followers', 'repositories']
for col in columns_to_fix:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```

```python
# Grouping by 'language' and summing the 'stars' to get the total stars per language
stars_by_language = df.groupby('language')['stars'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
stars_by_language.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("tab10", len(stars_by_language)), startangle=90, cmap="tab10")
plt.title("Distribution of Stars by Programming Language", fontsize=14)
plt.ylabel('')  # Hides the y-axis label
plt.show()
```

```python
# Convert 'followings' to numeric, forcing errors to NaN
df['followings'] = pd.to_numeric(df['followings'], errors='coerce')

# Drop rows where 'followings' is NaN (invalid data)
df = df.dropna(subset=['followings'])

# Get the top 10 rows with the highest followings
top_10_followings = df.nlargest(5, 'followings')

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the plot with enhanced design
plt.figure(figsize=(12, 7))
bar_plot = sns.barplot(x=top_10_followings['username'], y=top_10_followings['followings'], palette='magma')

# Add title and labels with improved fonts and sizes
plt.title("Top 10 Users by Followings", fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel("Username", fontsize=14, fontweight='bold', color='gray')
plt.ylabel("Followings", fontsize=14, fontweight='bold', color='gray')

# Add the values on top of each bar for clarity
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.0f'),  # Format the value to have no decimal places
                      (p.get_x() + p.get_width() / 2., p.get_height()),  # Positioning on top of the bar
                      ha='center', va='center', fontsize=12, color='black', fontweight='bold',
                      xytext=(0, 8), textcoords='offset points')  # Offset the label for better visibility

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=12)

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```

```python
top_languages = df['language'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_languages.index, y=top_languages.values, palette='magma')
plt.title("Top Languages Used by Users")
plt.xlabel("Language")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
# Repositories by language
repositories_by_language = df.groupby('language')['repositories'].sum().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(12, 6))
repositories_by_language.plot(kind='bar', color='skyblue')
plt.title("Number of Repositories by Language", fontsize=14)
plt.xlabel("Programming Language", fontsize=12)
plt.ylabel("Number of Repositories", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Distribution of stars
plt.figure(figsize=(10, 6))
sns.boxplot(x='language', y='stars', data=df, palette="coolwarm")
plt.title("Distribution of Stars by Language", fontsize=14)
plt.xlabel("Programming Language", fontsize=12)
plt.ylabel("Stars", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by language and year
language_trends = df.groupby(['joining_year', 'language'])['last_year_contributions'].sum().unstack().fillna(0)

# Visualization
plt.figure(figsize=(12, 6))
language_trends.plot(kind='line', colormap="viridis")
plt.title("Language Trends Over Time by User Activity", fontsize=14)
plt.xlabel("Joining Year", fontsize=12)
plt.ylabel("Total Contributions", fontsize=12)
plt.legend(title="Language", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['followers'], y=df['last_year_contributions'], size=df['repositories'], hue=df['language'], alpha=0.7, sizes=(50, 300))
plt.title("Activity Based on Follower Count and Number of Repositories", fontsize=14)
plt.xlabel("Followers", fontsize=12)
plt.ylabel("Last Year Contributions", fontsize=12)
plt.legend(title="Language", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Most popular repository by language
most_popular_repo = df.loc[df.groupby('language')['stars'].idxmax(), ['name', 'language', 'stars']]

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=most_popular_repo['stars'], y=most_popular_repo['language'], palette="mako")
plt.title("Most Popular Repositories by Stars and Language", fontsize=14)
plt.xlabel("Stars", fontsize=12)
plt.ylabel("Programming Language", fontsize=12)
plt.show()

# Display popular repositories
print(most_popular_repo)
```

```python
# Filtering users who joined most recently
most_recently_joined = df[df['sorting type'] == 'Most recently joined']
most_recently_joined = most_recently_joined.sort_values(by='joining_year', ascending=False)

# Plotting the most recently joined users
plt.figure(figsize=(10, 6))
sns.countplot(x='joining_year', data=most_recently_joined, palette="viridis")
plt.title("Most Recently Joined Users by Year", fontsize=14)
plt.xlabel("Joining Year", fontsize=12)
plt.ylabel("Number of Users", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering users with the most repositories
most_repositories = df[df['sorting type'] == 'Most repositories']
most_repositories = most_repositories.sort_values(by='repositories', ascending=False).head(10)

# Plotting the top 10 users with the most repositories
plt.figure(figsize=(10, 6))
sns.barplot(x='repositories', y='name', data=most_repositories, palette="Blues_d")
plt.title("Top 10 Users with Most Repositories", fontsize=14)
plt.xlabel("Number of Repositories", fontsize=12)
plt.ylabel("User Name", fontsize=12)
plt.show()
```

```python
# Filtering users with the most followers
most_followers = df[df['sorting type'] == 'Most followers']
most_followers = most_followers.sort_values(by='followers', ascending=False).head(10)

# Plotting the top 10 users with the most followers
plt.figure(figsize=(10, 6))
sns.barplot(x='followers', y='name', data=most_followers, palette="magma")
plt.title("Top 10 Users with Most Followers", fontsize=14)
plt.xlabel("Number of Followers", fontsize=12)
plt.ylabel("User Name", fontsize=12)
plt.show()
```

```python
# Filtering users with the fewest followers
fewest_followers = df[df['sorting type'] == 'Fewest followers']
fewest_followers = fewest_followers.sort_values(by='followers', ascending=True).head(10)

# Plotting the top 10 users with the fewest followers
plt.figure(figsize=(10, 6))
sns.barplot(x='followers', y='name', data=fewest_followers, palette="coolwarm")
plt.title("Top 10 Users with Fewest Followers", fontsize=14)
plt.xlabel("Number of Followers", fontsize=12)
plt.ylabel("User Name", fontsize=12)
plt.show()
```

```python
# Create a new column for the custom metric
df['best_match_score'] = df['followers'] + df['repositories']

# Filtering the top 10 best match users
best_match = df[df['sorting type'] == 'Best match']
best_match = best_match.sort_values(by='best_match_score', ascending=False).head(10)

# Plotting the top 10 best match users based on the custom score
plt.figure(figsize=(10, 6))
sns.barplot(x='best_match_score', y='name', data=best_match, palette="crest")
plt.title("Top 10 Best Match Users", fontsize=14)
plt.xlabel("Best Match Score (Followers + Repositories)", fontsize=12)
plt.ylabel("User Name", fontsize=12)
plt.show()
```

```python
# Filtering users with the fewest repositories
fewest_repositories = df[df['sorting type'] == 'Fewest repositories']
fewest_repositories = fewest_repositories.sort_values(by='repositories', ascending=True).head(10)

# Plotting the top 10 users with the fewest repositories
plt.figure(figsize=(10, 6))
sns.barplot(x='repositories', y='name', data=fewest_repositories, palette="BuGn")
plt.title("Top 10 Users with Fewest Repositories", fontsize=14)
plt.xlabel("Number of Repositories", fontsize=12)
plt.ylabel("User Name", fontsize=12)
plt.show()
```

```python
# Filtering users who joined the least recently
least_recently_joined = df[df['sorting type'] == 'Least recently joined']
least_recently_joined = least_recently_joined.sort_values(by='joining_year', ascending=True)

# Plotting the least recently joined users
plt.figure(figsize=(10, 6))
sns.countplot(x='joining_year', data=least_recently_joined, palette="viridis")
plt.title("Least Recently Joined Users by Year", fontsize=14)
plt.xlabel("Joining Year", fontsize=12)
plt.ylabel("Number of Users", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
